```python
class Book:
  def __init__(self, title, author, pages):
    self.title = title
    self.author = author
    self.pages = pages

  def get_info(self):
    print(f"Title : {self.title}")
    print(f"Author : {self.author}")
    print(f"Pages : {self.pages}")

obj1 = Book("Python Crash Course", "Eric Matthes", 544)
obj1.get_info()
obj2 = Book("Introduction to Machine Learning with Python", "Andreas C. Müller and Sarah Guido", 400)
obj2.get_info()
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

class Student(Person):
  def __init__(self, name, age, student_id):
    super().__init__(name, age)
    self.name = name
    self.age = age
    self.student_id = student_id

  def display_info(self):
    print(f"Student Name : {self.name}")
    print(f"Student Age : {self.age}")
    print(f"Student ID : {self.student_id}")

obj = Student("Anosha", 21, 123)
obj.display_info()
```

```python
class  StudentRecord:
  def __init__(self, student_name, score):
    self.student_name = student_name
    self.score = score

  def write_read(self):
    with open("record.txt", "a") as file:
      file.write(f"Student Name: {self.student_name}, Score: {self.score}\n")

    print(f"Student Name: {self.student_name}, Score: {self.score}")


obj1 = StudentRecord("Anosha", 10)
obj1.write_read()
obj2 = StudentRecord("Rabail", 9)
obj2.write_read()
obj3 = StudentRecord("Fatima", 10)
obj3.write_read()
```
